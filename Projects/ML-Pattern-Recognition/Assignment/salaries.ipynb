```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salaries.data')
```

```python
np_sal = df.to_numpy()
```

```python
np_ind = np.array((10,5))
np_usa = np.array((10,5))
```

```python
list_ind_sal = []
list_us_sal = []
list_ind_exp = []
list_us_exp = []
```

```python
for i in range(0,20):
    if np_sal[i][0] == 'India':
        list_ind_sal.append(np_sal[i][2])
        list_ind_exp.append(np_sal[i][4])
    else:
        list_us_sal.append(np_sal[i][2])
        list_us_exp.append(np_sal[i][4])
```

```python
plt.scatter(list_ind_exp,list_ind_sal,color='b')
plt.scatter(list_us_exp,list_us_sal,color='g')
plt.show()
#Blue - India
#Green - Abroad (USA, Canada, Russia)
```
